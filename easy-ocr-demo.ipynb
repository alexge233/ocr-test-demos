```python
import easyocr
from PIL import Image
import matplotlib.pyplot as plt
reader = easyocr.Reader(['en'])
```

```python
img = Image.open('test_image_2.png')
plt.imshow(img)
```

The Parameters `width_ths` and `height_ths` control how much overlap (width or height) of bounding boxes will allow a *merge* of words into a sentence.
A higher value requires a higher overlap, a lower value requires a lower overlap.
In this case the word `CHILD` is clearly overlapiong, hence why we end up with it in the results.
At the same time, we see that `maternity` and `Hospital` end up not included in the results.

```python
result = reader.readtext('test_image_2.png', width_ths=0.2, height_ths=0.2, decoder='beamsearch'), #slope_ths=0.1)
```
